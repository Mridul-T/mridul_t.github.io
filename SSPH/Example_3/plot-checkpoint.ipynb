```python
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 10))
def update_plot(frame):
    ax.clear()
    global coords
    coords_cpu = coords[0,:,:,frame]
    colors = ['red']

    
    ax.clear()
    ax.scatter(coords_cpu[:, 0], coords_cpu[:, 1], c=colors)
    ax.set_title(f"Scatter Plot of both boxes at time step {frame*2}")
    ax.set_xlabel("X coordinate")
    ax.set_ylabel("Y coordinate")
    
    
    ax.set_aspect('equal', adjustable='box')  # Set uniform scale
    ax.axvline(x=0, color='gray', linestyle='--')  # Add a vertical line at x = 0
    # plt.show()

anim = FuncAnimation(fig, update_plot, frames=coords.shape[3], interval=50)

anim.save('coords_pce.gif', writer=animation.PillowWriter(fps=60))

# Display the animation
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 10))
def update_plot(frame):
    ax.clear()
    global coords2
    coords_cpu = coords2[:,:,frame]
    colors = ['red']

    
    ax.clear()
    ax.scatter(coords_cpu[:, 0], coords_cpu[:, 1], c=colors)
    ax.set_title(f"Scatter Plot of both boxes at time step {frame*2}")
    ax.set_xlabel("X coordinate")
    ax.set_ylabel("Y coordinate")
    
    
    ax.set_aspect('equal', adjustable='box')  # Set uniform scale
    ax.axvline(x=0, color='gray', linestyle='--')  # Add a vertical line at x = 0
    # plt.show()

anim = FuncAnimation(fig, update_plot, frames=coords2.shape[2], interval=50)

anim.save('coords_det.gif', writer=animation.PillowWriter(fps=60))

# Display the animation
plt.show()
```
